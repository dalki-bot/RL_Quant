```python
window_size = 100
test_window_size = 10
def generate_random_df_slice(df, window_size, test_window_size):
    # 최대 시작 인덱스를 계산합니다. 이는 데이터프레임의 길이에서 윈도우 크기와 테스트 윈도우 크기를 뺀 값입니다.
    max_start_index = len(df) - window_size - test_window_size
    
    # 만약 최대 시작 인덱스가 0 이하라면, 데이터 크기가 너무 작아 분할할 수 없다는 에러를 발생시킵니다.
    if max_start_index <= 0:
        raise ValueError("데이터 크기가 너무 작아 분할할 수 없습니다.")

    # 시작 인덱스를 랜덤하게 선택합니다. 선택 범위는 0부터 최대 시작 인덱스입니다.
    start_index = np.random.randint(0, max_start_index)
    # 끝 인덱스는 시작 인덱스에 윈도우 크기를 더한 값입니다.
    end_index = start_index + window_size
    # 테스트 끝 인덱스는 끝 인덱스에 테스트 윈도우 크기를 더한 값입니다.
    test_end_index = end_index + test_window_size

    # 데이터프레임을 시작 인덱스와 테스트 끝 인덱스로 슬라이싱하고, 시작 인덱스와 끝 인덱스로 슬라이싱하고,
    # 끝 인덱스와 테스트 끝 인덱스로 슬라이싱한 결과를 반환합니다.
    return df[start_index:test_end_index], df[start_index:end_index], df[end_index:test_end_index]

def next_observation(self):
    # 현재 인덱스를 기준으로 이전 100개의 관찰값을 슬라이싱하여 반환합니다.
    observation_df = self.slice_df.iloc[(self.current_index-101)+1:(self.current_index-1)+1]
    return observation_df
```
